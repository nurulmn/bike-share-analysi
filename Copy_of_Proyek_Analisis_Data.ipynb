### Cleaning Data

```python
df_day['dteday'] = pd.to_datetime(df_day['dteday'])
df_hour['dteday'] = pd.to_datetime(df_hour['dteday'])
df_day.rename(columns={'yr': 'year', 'mnth': 'month'}, inplace=True)
df_hour.rename(columns={'yr': 'year', 'mnth': 'month', 'hr': 'hour'}, inplace=True)
df_day['cnt_check'] = df_day['casual'] + df_day['registered']
df_hour['cnt_check'] = df_hour['casual'] + df_hour['registered']
```

```python
df_day['hum'].replace(0, pd.NA, inplace=True)
df_hour['hum'].replace(0, pd.NA, inplace=True)
df_day['hum'].fillna(df_day['hum'].median(), inplace=True)
df_hour['hum'].fillna(df_hour['hum'].median(), inplace=True)
```

```python
# 5. Memeriksa dan menangani anomali pada kolom 'windspeed'
# Mengganti nilai 0 dengan nilai NaN agar dapat diisi atau dihapus sesuai kebijakan
df_day['windspeed'].replace(0, pd.NA, inplace=True)
df_hour['windspeed'].replace(0, pd.NA, inplace=True)

# Mengisi nilai NaN pada 'windspeed' dengan median atau nilai yang sesuai
df_day['windspeed'].fillna(df_day['windspeed'].median(), inplace=True)
df_hour['windspeed'].fillna(df_hour['windspeed'].median(), inplace=True)
```

```python
df_day.drop(columns=['instant'], inplace=True)
df_hour.drop(columns=['instant'], inplace=True)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pengaruh musim terhadap jumlah penggunaan sepeda harian?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='season', y='cnt', data=df_day)
plt.title('Distribusi Pengguna Berdasarkan Musim')
plt.xlabel('Musim (1: Musim Semi, 2: Musim Panas, 3: Musim Gugur, 4: Musim Dingin)')
plt.ylabel('Total Pengguna')
plt.show()
```

Penggunaan sepeda lebih rendah pada musim semi (Musim 1), dengan median yang terlihat lebih rendah dibandingkan musim lainnya.
Musim panas (Musim 2) dan musim gugur (Musim 3) memiliki jumlah pengguna sepeda yang lebih tinggi, dengan median di atas 5000 pengguna.
Musim dingin (Musim 4) menunjukkan penurunan jumlah pengguna sepeda dibandingkan dengan musim panas dan gugur, tetapi masih lebih tinggi daripada musim semi.

### Pertanyaan 2: Apakah ada pengaruh Hari Kerja dan Akhir Pekan?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='workingday', y='cnt', data=df_day)
plt.title('Pengguna Berdasarkan Hari Kerja vs Akhir Pekan')
plt.xlabel('Hari Kerja (0: Akhir Pekan/Hari Libur, 1: Hari Kerja)')
plt.ylabel('Total Pengguna')
plt.show()
```

Observasi:
Tidak ada perbedaan yang signifikan antara jumlah pengguna sepeda pada hari kerja dan akhir pekan/hari libur, dengan median yang hampir sama.
Namun, ada sedikit lebih banyak variasi pada hari kerja, terlihat dari rentang interkuartil yang sedikit lebih lebar.

## Conclusion

- Conclution pertanyaan 1
Musim panas dan musim gugur adalah musim yang paling populer untuk penggunaan sepeda, kemungkinan karena cuaca yang lebih mendukung untuk aktivitas luar ruangan.
Musim semi memiliki jumlah pengguna yang paling rendah, kemungkinan karena cuaca yang lebih tidak menentu atau suhu yang lebih rendah.
